# Data Visualization with Matplotlib and Seaborn

`Matplotlib` and `Seaborn` are Python libraries for creating and customizing various data visualizations. `Matplotlib` provides a versatile foundation, while `Seaborn` simplifies complex statistical visualizations.

## Introduction to `Matplotlib`
`Matplotlib` is a comprehensive library for creating static, animated, and interactive visualizations in Python. `Matplotlib` is important for data visualization as it
helps in understanding the trends, patterns, and relationships
in the data. With `Matplotlib`, you can create various types of plots, including line plots, bar charts, scatter plots, histograms, and more. It offers precise control over every aspect of your plots, allowing you to customize colors, labels, axes, and annotations to suit your specific needs. Let's get started!

## Creating a line plot

```python
import numpy as np

import matplotlib.pyplot as plt


x = np.arange(1,8)
y = np.square(x) # or np.power(x,2) or x**2

plt.plot(x,y)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Simple Line Plot")
plt.show()
```

### Elements of the line plot
`plt.plot(x,y)` is the command that creates the actual line plot. It takes the data from x and y and plots it on the graph.

`plt.xlabel('X-axis')` is used to add a label to the X-axis of the plot, specifying the axis's name.

`plt.ylabel('Y-axis')` adds a label to the Y-axis, indicating what the Y-axis represents.

`plt.title('Simple Line Plot')` sets the title of the plot, which provides context or a brief description of the data being visualized.

`plt.show()` Finally, the show function is called to display the plot. Without this line, the plot won't be visible on the screen.

### Customize your plot
1. The appearance of your plot can be customized with `color`, `linestyle` and `linewidth` arguments.
2. You can use the `label` argument to identify your plot, and then `plt.legend()` to show the various labels.
3. `plt.xlim` and `plt.ylim` constrain the figure to the given values.
4. `plt.grid()` can be used to add grid lines to your plot, and `plt.axhline` and `plt.axvline` draw horizontal and vertical lines at required values respectively.



```python
x = np.linspace(0,2*np.pi,10)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1, color='red', linewidth='2', linestyle='solid', label='Cosine')
plt.plot(x, y2, color='green', linewidth='4', linestyle='--', label='Sine')

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Sine and Cosine")

plt.grid()

plt.legend()

# plt.ylim(0,0.5) # ytour comments
# plt.xlim(0,0.25)

plt.axhline(y=1)
plt.axvline(x=np.pi/2.0)
plt.show()
```

### Creating multiple subplots in one figure

`plt.subplots(row_dim, col_dim`) will return a `fig` and an `ax` object. The `ax` object can be referenced to create multiple sub-plots within the same `fig`



```python
x = np.linspace(0,2*np.pi,1000)
y1 = np.cos(x)
y2 = np.sin(x)

fig, ax = plt.subplots(1,2)
ax[0].plot(x,y1)

ax[1].plot(x,y2)


ax[0].set_title("Cosine")
ax[1].set_title("Sine")

plt.suptitle("Sine and Cosine")
```

### Add final touches and save your plot
`plt.xticks` and `plt.yticks` let you further customize your axes.
Use `plt.savefig` to save your figure in an appropriate format.

```python
x = np.linspace(0,2*np.pi,1000)
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x, y1, color='red', linewidth='2', linestyle='-', label='Cosine')
plt.plot(x, y2, color='green', linewidth='4', linestyle='dotted', label='Sine')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Sine and Cosine")
plt.grid()
plt.legend()
plt.ylim()
plt.axhline(y=1,color ='k', linestyle='--')
plt.axvline(x=np.pi/2.0)

plt.xticks([0,np.pi/2.0,np.pi,3*np.pi/2.0,2*np.pi], labels = ['Zero','','Pi','','2 Pi'])
plt.yticks([-1.0,-0.5,0.0,0.5,1.0])

plt.savefig("SineCosine.png")
plt.show()
```

## Some more plots with Matplotlib
You can use matplotlib to generate several types of plots. Given below are two examples, a bar plot and a scatter plot.

`plt.bar` generates bar plots. You can create both stacked and staggered bar plots for comparing multiple datasets.


```python
# Bar Plot
plt.figure(figsize=(11,8))

cells = ['1x1', '3x3', '5x5']

sota_speed = [43.2,5.1,0.54]
our_speed = [117.3,69.4,30.6]
# Create a bar plot
X_axis = np.arange(len(cells))

plt.bar(X_axis - 0.2, sota_speed, 0.4, label = 'SOTA Code', zorder=3)
plt.bar(X_axis + 0.2, our_speed, 0.4, label = 'Our Code', zorder=3)

# Customize the plot
plt.xticks(X_axis, cells, size=20)
plt.yticks(size=20)
plt.xlabel("Domain Size", size=20)
plt.ylabel("Mean Epoch/s", size=20)
plt.title("Performance Comparison", size=20)
plt.legend(fontsize="20")

plt.grid(axis='y', alpha=0.5, zorder=0)


# Save the plot
plt.savefig("BarPlotStaggered.png",dpi=600 )
```

```python
# Bar Plot
plt.figure(figsize=(11,8))

cells = ['1x1', '3x3', '5x5']
sota_speed = [43.2,5.1,0.54]
our_speed = [117.3,69.4,30.6]
# Create a bar plot
X_axis = np.arange(len(cells))

plt.bar(X_axis , sota_speed, 0.4, label = 'SOTA Code')#, zorder=3)
plt.bar(X_axis , our_speed, 0.4, bottom=sota_speed, label = 'Our Code')#, zorder=3)

# Customize the plot
plt.xticks(X_axis, cells, size=20)
plt.yticks(size=20)
plt.xlabel("Domain Size", size=20)
plt.ylabel("Mean Epoch/s", size=20)
plt.title("Performance Comparison", size=20)
plt.legend(fontsize="20")

# plt.grid(axis='y', alpha=0.5, zorder=0)


# Save the plot
plt.savefig("BarPlotStacked.png",dpi=600 )
```

### Scatter Plots
Use `plt.scatter` to visualize the distribution of your data points.

```python
# Scatter plot
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o', s=100)
plt.scatter(x+.5,z+1,color='red',marker='x',s=50) # Create some boundary between the two datasets by shifting x and z

# Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')

# Save the plot
plt.savefig("Scatter.png")

# Display the plot
plt.show()
```

### Contour Plots
`plt.contour` and `plt.contourf` can be used to create contours and filled contours respectively.

`plt.figure(figsize=(x_size, y_size))` sets the figure size in inches.

```python
x = np.linspace(-1,1,1000) # 1d arrays
y = np.linspace(-1,1,1000) #1 d arrays
xx, yy = np.meshgrid(x,y)
plt.figure(figsize=(4,4))
plt.contour(xx,yy,(xx**2+yy**2))
```

The `levels` keyword controls how fine the contours will be. Additionally `plt.colorbar` will display a the color to value map as a color bar on the side of the image.

```python
x = np.linspace(-1,1,1000)
y = np.linspace(-1,1,1000)
xx, yy = np.meshgrid(x,y)
plt.figure(figsize=(10,8))
plt.contourf(xx,yy,((xx**2)/2+(yy**2)/3), levels=100, cmap="jet")
plt.colorbar()
```

## Introduction to `Seaborn`
`Seaborn` is a Python data visualization library based on `Matplotlib`. It provides a high-level interface for drawing attractive
and informative statistical graphics. It is built on top of `Matplotlib` and closely integrated with `pandas` data structures. `Seaborn` comes with some in-built dataframes that can be loaded using `load_dataset`

```python
# importing packages
import seaborn as sns

# loading dataset
data_titanic = sns.load_dataset("titanic")
data_tips = sns.load_dataset("tips")
data_flights = sns.load_dataset("flights")

```

```python
data_flights.head()
```

### Line Plots
Dataframes can be directly passed to `Seaborn`, and the x and y axes can be specified using column names.

```python
data_flights_month = data_flights[data_flights["month"]=="Jan"]

sns.lineplot(x="year",y="passengers",data=data_flights_month)
```

In addition, `Seaborn` offers several advanced features, like the ability to generate multiple plots from the same data using `hue`

```python
sns.lineplot(x="year",y="passengers",data=data_flights, hue="month")
plt.grid()
plt.title("Flights")
```

### Scatter Plots
Scatter plots in `seaborn` come with a variety of high-level features like `hue`, `style` and `size`

```python
data_tips.head()
```

```python
sns.scatterplot(data=data_tips, x="total_bill", y="tip",hue="day",style="time",size="sex")
plt.grid()
```

### Statistical Visualization - `boxplot`, `violinplot` and `histplot`
Seaborn includes some excellent tools for visualizing statistics of your dataset.

```python
data_titanic.head()
```

```python
sns.boxplot(data=data_titanic, x="class", y="age", hue="alive")
```

```python
sns.violinplot(data=data_titanic, x="age", y="class") #check with hue
```

```python
sns.histplot(data=data_titanic,x="age",bins=10) #
```

```python
sns.kdeplot(data=data_titanic,x="age", hue="alive")
```

### Heatmap
`sns.heatmap` is particularly useful for visualizing data in a matrix form, for example correlation matrices.

```python
group = data_titanic[data_titanic["alive"]=="yes"].groupby(["class", "sex"])
data_alive = group.size().unstack()
data_alive
```

```python
df_alive = data_titanic[data_titanic["alive"]=="yes"]
df_alive.head()
```

```python
df_grouped = df_alive.groupby(["class", "sex"])
size_series = df_grouped.size()
```

```python
size_series
```

```python
data_alive = size_series.unstack() #dtype?
```

```python
sns.heatmap(data_alive,annot=True,cbar=True)
```

## Creating subplots with `FacetGrid`
`FacetGrid` conveniently reads the datafram and the variables you want to plot in the rows and columns, and then creates the appropriate number of subplots. Any `seaborn` function can then be mapped on to this array of subplots.

```python
data_tips.head()
```

```python
g = sns.FacetGrid(data=data_tips, col="time", row="sex", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
```
